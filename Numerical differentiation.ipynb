```python
import numpy as np
import math
```

```python
def f(x) :
  #return x**2
  #return np.exp(x)*(np.sin(x)/x)
  return x*np.exp(x)
```

```python
x = 2
#x = 0.5

h = 0.05

df_forward = (f(x+h)-f(x))/h #Forward difference formula

df_backward = (f(x)-f(x-h))/h #Backward difference formula

df_central = (f(x+h)-f(x-h))/(2*h) #Central difference formula

print("Answer for forward difference =",df_forward)
print("Answer for backward difference =",df_backward)
print("Answer for central difference =",df_central)
```

```python
x = 2

h = 0.5

d2f = (f(x+h)+f(x-h)-2*f(x))/(h**2) #Second derivative approximation

print("The second derivative of the given f(x) is",d2f)
```
