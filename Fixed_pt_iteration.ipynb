```python
import math #Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g1(x):
    return math.sqrt(2*x)
```

```python
def g2(x):
    return ((x**2)/2)
```

```python
x0 = 2.5
tol = 0.01
error=1
root = np.empty(0)
iterations = 0

while (error>tol):
#for i in range(100):
    #x1=g1(x0);
    x1=g2(x0);
    iterations +=1
    if iterations>10:
      print("The given g(x) does not converge")
      break
    error = abs(x1-x0)
    root = np.append(root,error)
    x0=x1

print (x0)
```
