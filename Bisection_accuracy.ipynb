```python
def f(x): #To find the value of f(x) given in the problem
    return 4*(x**2)+3*x-3
```

```python
a=0
b=1
#tol=0.01
error=abs(a-b)
c=0
#err_arr=np.empty(0)

n=30
p = int(input("Enter order of accuracy="))

#while(error>tol):
for i in range(n):
    prev_root = c # Value from previous iteration

    c=(a+b)/2 # Bisection method
    print(c)

    if((f(a)*f(c))<0): # Condition for bisecting the limit
        b = c # Discard old value of b
    else:
        a = c # Discard old value of a

    current_root = c
    accuracy = abs(current_root-prev_root)
    if (accuracy < 10**-p):
       print ("The root is ", c)
       break
```
