```python
import numpy as np

# Initialize matrix A and vector x
A = np.array([[1,0,0,0], [6,2,0,0], [7,4,3,0], [5,6,7,-4]])
#A = np.array([[3,0,0], [-4,6,2], [16,-15,-5]])
#A = np.array([[2,-1],[-1,1]])

# Set the initial guess to [1, 1]
x = np.ones((4, 1))  # Column vector

# Initial values for the loop
xmax1 = 1000
xmax2 = 1
iterations = 0
error = abs(xmax1 - xmax2)
tol = 0.001

# Iterative process
while error > 0.001:
    iterations += 1
    x1 = A @ x  # Matrix multiplication
    xmax2 = np.max(np.abs(x1))  # Find the maximum absolute value

    I = np.argmax(np.abs(x1))  # Index of the maximum value
    xmax2 = x1[I]   # Update eigenvalue estimate

    x1 = x1 / xmax2  # Normalize the vector
    error = abs(xmax2 - xmax1)  # Calculate error

    x = x1  # Update vector x
    xmax1 = xmax2  # Update previous eigenvalue

# Print results
print("Dominant Eigenvalue:", xmax2)
print("Dominant Eigenvector:\n", x)
print("Iterations:", iterations)
```
