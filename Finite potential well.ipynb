```python
#Importing required libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
m_e = 9.109384e-31
eV = 1.602176e-19
h_bar = 1.054572e-34

hbar2 = h_bar**2 * (1/m_e) * (1/eV) * (1e18)
print(hbar2)
```

```python
hcross_factor = 2/hbar2
hcross_factor
```

```python
def f(V0, a, E):
    term1 = np.sqrt(hcross_factor*(V0-E))
    term2 = np.sqrt(hcross_factor*E)*np.tan(np.sqrt(hcross_factor*E)*a)
    return term1-term2
```

```python
def df(V0, a , E):
    term3 = 0.5*(hcross_factor*(V0-E))**(-0.5)*(-hcross_factor)
    term4 = 0.5*(hcross_factor*E)**(-0.5)*(hcross_factor)*np.tan(np.sqrt(hcross_factor*E)*a)

    x = np.sqrt(hcross_factor*E)*a
    sec_x = 1/np.cos(x)
    secx_sqr = sec_x**2
    term5 = np.sqrt(hcross_factor*E)*secx_sqr*0.5*(hcross_factor*E)**(-0.5)*a*hcross_factor

    return term3-term4-term5
```

```python
E0 = float(input("Enter the starting value ="))

V0 = 10
a = 0.3

n=20
tol=0.01
error = E0

#while(error>tol):
for i in range (n):
  E1 = E0 - (f(V0,a,E0)/df(V0,a,E0))
  print(E1)
  error = abs(E1-E0)

  E0=E1

print("The root is =", E1)
```
