```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f(x) :
  #return x**2
  return np.exp(x)*(np.sin(x)/x)
  #return x*np.exp(x)
```

```python
#x = 2
x = 0.5
df = np.exp(x)*(np.sin(x)/x)+ np.exp(x)*(np.cos(x)/x)- np.exp(x)*(np.sin(x)/x**2)

h = [0.45, 0.1, 0.05, 0.01, 0.005]

err_forward = np.empty(0)
err_backward = np.empty(0)
err_central = np.empty(0)

for i in range(len(h)):
  df_forward = (f(x+h[i])-f(x))/h[i] #Forward difference formula
  #print("Answer for forward difference =",df_forward)
  err_forward = np.append(err_forward,(abs(df-df_forward)))

  df_backward = (f(x)-f(x-h[i]))/h[i] #Backward difference formula
  err_backward = np.append(err_backward,(abs(df-df_backward)))

  df_central = (f(x+h[i])-f(x-h[i]))/(2*h[i]) #Central difference formula
  print("Answer for central difference =",df_central)
  err_central = np.append(err_central,(abs(df-df_central)))
```

```python
print(err_forward)
print(err_backward)
print(err_central)
```

```python
plt.plot(err_forward)
plt.plot(err_backward)
plt.plot(err_central)
plt.legend(['Forward', 'Backward', 'Central'])
```
