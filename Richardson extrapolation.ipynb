```python
import numpy as np
import math
```

```python
def f(x):
  return 5*x*np.exp(-2*x)
```

```python
def D(x,h):
  return (f(x+h)-f(x-h))/(2*h)
```

```python
x = 0.35

h = 0.125

D_h = D(x,h)
print(D_h)
D_2h = D(x,2*h)
D_4h = D(x,4*h)
D_8h = D(x,8*h)

D1_h_2h = (4*D_h - D_2h)/(2**2-1)
print(D1_h_2h)
D1_2h_4h = (4*D_2h - D_4h)/(2**2-1)
D1_4h_8h = (4*D_4h - D_8h)/(2**2-1)

D2_4h = (2**4*D1_h_2h - D1_2h_4h)/(2**4-1)
print(D2_4h)
D2_8h = (2**4*D1_2h_4h - D1_4h_8h)/(2**4-1)

D3 = (2**6*D2_4h - D2_8h)/(2**6-1)
D3
```
