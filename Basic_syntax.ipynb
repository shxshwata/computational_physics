```python
#Importing required libraries
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
for i in range(10): # For loop using range(n)
    print(i)
```

```python
for i in range(5, 15): # range(start, stop)
    print(i)
```

```python
start = int(input ("Enter starting value =")) # Getting input values from user
stop = int(input ("Enter stopping value ="))
step =int(input ("Enter increment step ="))
for i in range(start, stop, step): # range(start, stop, step)
    print(i)
```

```python
for i in range(1, 4):  # Nested for loops
    for j in range(1, 4):
        print(i, j)
```

```python
age = 28
count = 0 # counter

# the test condition is always True
while age > 19:
    count +=1
    if(count > 3):
        print('Infinite Loop')
        break
    print(count)
```

```python
# Plotting graphs

# setting the x - coordinates
x = np.arange(0, 2*(np.pi), 0.1)
# setting the corresponding y - coordinates
y = np.sin(x)
z = np.cos(x)

# plotting the points
plt.plot(x, y, color = "green")
plt.plot(x, z, color = "purple")

# function to show the plot
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
y = 4*(x**2)+3*x-3

plt.plot(x,y)
plt.axhline(y = 0, color="black",linestyle="--")
plt.grid(True)
plt.show()
```
