Program to compare the accuracy of Lagrangian interpolation and series expansion in approximating exp(x)

```python
import numpy as np
```

```python
n = int(input("Enter the number of points to be used ="))
```

```python
x = np.empty(0)
fx = np.empty(0)

for i in range(n):
  a = float(input(f"Enter point number {i+1}: "))
  x = np.append(x,a)
  b = float(input(f"Enter value at point number {i+1}: "))
  fx = np.append(fx,b)
```

```python
xi = float(input("Enter the point to be evaluated ="))
sumf = 0

for i in range(n):
  num = 1
  denom = 1
  for j in range(n):
    if i!=j:
      num = num*(xi-x[j])
      denom = denom*(x[i]-x[j])
    else:
      num = fx[i]*num
  sumf = sumf+(num/denom)

print(sumf)
```
