```python
import numpy as np
```

```python
A = np.array([   #Enter the matrix to be solved
    [1, 2, 3],
    [4, 5, 6],
    [3, 1, -2]])

# RHS vector
b = np.array([9, 24, 4])

# Number of unknowns
n = 3

# Gaussian elimination method
for k in range(n-1):
    for i in range(k+1, n):
        factor = A[i, k]/A[k, k]
        b[i] = b[i] - factor*b[k]
        for j in range(k, n):
            A[i, j] = A[i, j] - factor*A[k, j]

# Back substitution
x = np.zeros(n)
x[n-1] = b[n-1]/A[n-1, n-1]
for i in range(n-2, -1, -1):
    sum = b[i]
    for j in range(i+1, n):
        sum = sum - A[i, j]*x[j]
    x[i] = sum/A[i, i]

print(x)
```
