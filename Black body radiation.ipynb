```python
import numpy as np
```

```python
iterations = 20
x0 = float(input("Enter the starting value ="))
x1=x0

for i in range(iterations):
  f_x = (5*(np.exp(x0)-1))-(x0*(np.exp(x0)))
  df_x = (4*(np.exp(x0)))-(x0*(np.exp(x0)))

  x1 = x0-((f_x)/(df_x))
  print(x1)

  x0=x1

print("The root is =", x1)
```
