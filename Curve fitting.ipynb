```python
import numpy as np
import matplotlib.pyplot as plt

#Step 1: Initializing the data
x = np.array([19.1, 25.0, 30.1, 36.0, 40.0, 45.1, 50.0])
y = np.array([76.30, 77.80, 79.75, 80.80, 82.35, 83.90, 85.10])

#x=np.array([2,4,6,8])
#y=np.array([10,16,22,28])

# Step 2: Calculate means
x_mean = np.mean(x)
y_mean = np.mean(y)

# Step 3: Calculate the slope (m)
numerator = np.sum((x - x_mean) * (y))
denominator = np.sum((x - x_mean) ** 2)
m = numerator / denominator

# Step 4: Calculate the intercept (c)
c = y_mean - m * x_mean

y_fit = m * x + c

print("y =",m,"x +",c)
```

```python
plt.scatter(x,y, color='green')
plt.plot(x,y_fit,"--")
plt.scatter(x_mean,y_mean, color='red', marker="^")
```
